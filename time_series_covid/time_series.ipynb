```python
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA

# Step 1: Load data
df = pd.read_csv("covid_19_india.csv")

# Step 2: Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Step 3: Group by Date to get total confirmed cases across the country
daily_cases = df.groupby('Date')['Confirmed'].sum()

# Step 4: Plot original time series
plt.figure(figsize=(10, 4))
plt.plot(daily_cases, label='Confirmed Cases')
plt.title("COVID-19 Confirmed Cases in India")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.legend()
plt.show()

# Step 5: Use auto_arima to find best model parameters
model_auto = auto_arima(daily_cases, seasonal=False, trace=True, suppress_warnings=True)

# Step 6: Fit ARIMA model with best parameters
model = ARIMA(daily_cases, order=model_auto.order)
model_fit = model.fit()

# Step 7: Forecast the next 30 days
forecast = model_fit.forecast(steps=30)

# Step 8: Plot the forecast
plt.figure(figsize=(10, 4))
plt.plot(daily_cases, label="Historical")
plt.plot(pd.date_range(daily_cases.index[-1], periods=31, freq='D')[1:], forecast, label="Forecast", color='red')
plt.title("ARIMA Forecast for COVID-19 Cases")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.legend()
plt.show()
```
